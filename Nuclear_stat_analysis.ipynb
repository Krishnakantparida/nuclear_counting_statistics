```python
import math
import matplotlib.pyplot as plt
import statistics as stat

list_A1 = []
with open("background count stat.txt") as A:
    for a in A:
        list_A1.append(int(a))

#print(list_A1)
list_B1 = [i for i in range(2,max(list_A1)+1)]
#print(list_B1)
print(stat.mean(list_A1), stat.stdev(list_A1),stat.variance(list_A1))
def histogram_maker(list_dist):
    plt.hist(list_dist,bins=11,ec='blue',density='true',label='Dist. Histogram')
    plt.show

def poisson_distribution(list_B1,list_A1):
    mean = stat.mean(list_A1)
    list_C1 = [(mean**k)*(math.exp(-mean))*(1/math.factorial(k)) for k in list_B1]
    plt.plot(list_B1,list_C1,'k--o', label='Poisson dist.')
    plt.legend(),plt.show()

plt.xlabel('Count values occured')
plt.ylabel('Frequency density distribution')
histogram_maker(list_A1)
poisson_distribution(list_B1,list_A1)
```

```python
import math
import matplotlib.pyplot as plt
import statistics as stat

list_A2 = []
with open("count stat with Co_60.txt") as A:
    for a in A:
        list_A2.append(int(a))

#print(list_A2)
list_B2 = [i for i in range(min(list_A2),max(list_A2)+1)]
#print(list_B2)

print(stat.mean(list_A2), stat.stdev(list_A2),stat.variance(list_A2))
def histogram_maker(list_dist):
    plt.hist(list_dist,bins=10,color='green',ec='darkgreen',density='true',label='Dist. Histogram')
    plt.show

def gaussian_distribution(list_B,list_A):
    mean, var = stat.mean(list_A), stat.variance(list_A)
    list_C = [((2*math.pi*var)**(-0.5))*(math.exp(-(k-mean)**2/(2*var))) for k in list_B]
    plt.plot(list_B,list_C,'k--',label='Gaussian Dist.')
    plt.legend(),plt.show()


plt.xlabel('Count values occured')
plt.ylabel('Frequency density distribution')
histogram_maker(list_A2)
gaussian_distribution(list_B2,list_A2)
```
